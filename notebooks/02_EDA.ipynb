```python
import pandas as pd
df = pd.read_csv("../data/processed/ekstraklasa_clean.csv")
df.head()
```

```python
# Dodanie podziału na grupy (Top 4, Mid, Bottom 3)
def tier(rk):
    if rk <= 4:
        return "Top 4"
    elif rk >= 16:
        return "Bottom 3"
    else:
        return "Mid"

df['Tier'] = df['Rk'].apply(tier)
df['Tier'].value_counts()

```

```python
# Porównanie statystyk między grupami
metrics = [
    'Pts', 'GF', 'GA', 'GD', 'Poss',
    'Sh/90', 'SoT/90', 'G/Sh', 'G/SoT'
]

df.groupby('Tier')[metrics].mean().round(2)
```

```python
# Wykres: Pts vs Poss
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(df['Poss'], df['Pts'])

for _, row in df.iterrows():
    plt.text(row['Poss'] + 0.1, row['Pts'] + 0.1, row['Squad'], fontsize=8)

plt.xlabel("Posiadanie piłki (%)")
plt.ylabel("Punkty")
plt.title("Zależność: posiadanie vs punkty – Ekstraklasa 24/25")
plt.grid(True)
plt.show()
```

```python
# które statystyki najbardziej wpływają na punkty
import seaborn as sns
import matplotlib.pyplot as plt

corr = df[['Pts','GF','GA','GD','Poss','Sh/90','SoT/90','G/Sh','G/SoT']].corr()

plt.figure(figsize=(10,7))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Macierz korelacji — Ekstraklasa 24/25")
plt.show()
```

```python
# NAJWAŻNIEJSZE czynniki sukcesu
corr['Pts'].sort_values(ascending=False)
```

```python
# Która drużyna jest najskuteczniejsza
df[['Squad','G/Sh','G/SoT']].sort_values(by='G/Sh', ascending=False)
```

```python
# Wykorzystanie strzałów
df['Efficiency'] = df['Gls_shoot'] / df['Sh']
df[['Squad','Efficiency']].sort_values(by='Efficiency', ascending=False)
```

```python
#Zależność: skuteczność vs punkty
plt.figure(figsize=(8,6))
plt.scatter(df['G/Sh'], df['Pts'])

for _,row in df.iterrows():
    plt.text(row['G/Sh']+0.001, row['Pts']+0.4, row['Squad'], fontsize=8)

plt.xlabel("Gole na strzał (G/Sh)")
plt.ylabel("Punkty")
plt.title("Skuteczność strzałów vs punkty")
plt.grid(True)
plt.show()
```

```python
# Zależność: skuteczność strzałów vs posiadanie
plt.figure(figsize=(8,6))
plt.scatter(df['Poss'], df['G/Sh'])

for _,row in df.iterrows():
    plt.text(row['Poss']+0.1, row['G/Sh']+0.001, row['Squad'], fontsize=8)

plt.xlabel("Posiadanie (%)")
plt.ylabel("Skuteczność strzałów (G/Sh)")
plt.title("Posiadanie piłki vs skuteczność strzałów")
plt.grid(True)
plt.show()
```

```python
# Radar chart (TOP4 vs BOTTOM3)
import numpy as np
import matplotlib.pyplot as plt

# Średnie dla grup
top4 = df[df['Tier'] == 'Top 4'][['Poss','Sh/90','SoT/90','G/Sh','G/SoT','GD']].mean()
bottom3 = df[df['Tier'] == 'Bottom 3'][['Poss','Sh/90','SoT/90','G/Sh','G/SoT','GD']].mean()

# Kategorie
categories = ['Poss', 'Sh/90', 'SoT/90', 'G/Sh', 'G/SoT', 'GD']
N = len(categories)

# Konwersja do list + zamknięcie wykresu (pierwsza wartość = ostatnia)
top4_vals = top4.tolist()
bottom3_vals = bottom3.tolist()

top4_vals += top4_vals[:1]
bottom3_vals += bottom3_vals[:1]

# Kąty
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
angles += angles[:1]

# Wykres
plt.figure(figsize=(8,8))
plt.subplot(111, polar=True)

# Top4
plt.plot(angles, top4_vals, label='Top 4', linewidth=2, color='blue')
plt.fill(angles, top4_vals, alpha=0.25, color='blue')

# Bottom3
plt.plot(angles, bottom3_vals, label='Bottom 3', linewidth=2, color='red')
plt.fill(angles, bottom3_vals, alpha=0.25, color='red')

# Ustawienia
plt.xticks(angles[:-1], categories, fontsize=12)
plt.title("Profil statystyczny — Top4 vs Bottom3 (Ekstraklasa 24/25)", fontsize=14)
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.1))
plt.show()
```

# Charakterystyki drużyn: Lech Poznań vs Widzew Łódź

```python
lech = df[df['Squad'] == 'Lech Poznań']
widzew = df[df['Squad'] == 'Widzew Łódź']

# wybieramy metryki do porównania
stats = ['Poss','Sh/90','SoT/90','G/Sh','G/SoT','GF','GA','GD','Pts']

comparison = pd.DataFrame({
    'Statystyka': stats,
    'Lech Poznań': lech[stats].iloc[0].values,
    'Widzew Łódź': widzew[stats].iloc[0].values
})

comparison
```

```python
# Wykres RADAR (Lech vs Widzew)
import numpy as np
import matplotlib.pyplot as plt

# wartości
lech_vals = lech[['Poss','Sh/90','SoT/90','G/Sh','G/SoT','GD']].iloc[0].values.tolist()
widzew_vals = widzew[['Poss','Sh/90','SoT/90','G/Sh','G/SoT','GD']].iloc[0].values.tolist()

categories = ['Poss','Sh/90','SoT/90','G/Sh','G/SoT','GD']
N = len(categories)

# domknięcie wykresu
lech_plot = lech_vals + lech_vals[:1]
widzew_plot = widzew_vals + widzew_vals[:1]

angles = np.linspace(0, 2*np.pi, N, endpoint=False).tolist()
angles += angles[:1]

plt.figure(figsize=(8,8))
plt.subplot(111, polar=True)

# Lech
plt.plot(angles, lech_plot, label='Lech Poznań', linewidth=2, color='blue')
plt.fill(angles, lech_plot, alpha=0.25, color='blue')

# Widzew
plt.plot(angles, widzew_plot, label='Widzew Łódź', linewidth=2, color='red')
plt.fill(angles, widzew_plot, alpha=0.25, color='red')

plt.xticks(angles[:-1], categories, fontsize=12)
plt.title("Profil statystyczny: Lech Poznań vs Widzew Łódź", fontsize=14)
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.1))

plt.show()
```

# Wykres Barplot

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8-whitegrid")  # ładny, neutralny styl
sns.set(font_scale=0.9)
```

```python
# Ranking ofensywy - gole strzelone
df_offense = df[['Squad', 'GF']].sort_values('GF', ascending=True)  # ascending=True, bo zrobimy poziomy wykres

plt.figure(figsize=(8,6))
sns.barplot(
    data=df_offense,
    x='GF',
    y='Squad',
    hue='Squad',
    palette='Blues_r',
    legend=False
)

plt.title("Gole strzelone (GF) – ranking drużyn Ekstraklasy 24/25")
plt.xlabel("Gole strzelone")
plt.ylabel("Drużyna")

# dodanie wartości na słupkach
for i, v in enumerate(df_offense['GF']):
    plt.text(v + 0.3, i, str(v), va='center')

plt.tight_layout()
plt.show()
```

```python
# Ranking defensywy - gole stracone
df_defense = df[['Squad', 'GA']].sort_values('GA', ascending=True)  # od najlepszej obrony

plt.figure(figsize=(8,6))
sns.barplot(
    data=df_defense,
    x='GA',
    y='Squad',
    hue='Squad',
    palette='Greens',
    legend=False
)

plt.title("Gole stracone (GA) – im mniej, tym lepiej")
plt.xlabel("Gole stracone")
plt.ylabel("Drużyna")

for i, v in enumerate(df_defense['GA']):
    plt.text(v + 0.3, i, str(v), va='center')

plt.tight_layout()
plt.show()
```

```python
# Ranking różnicy bramek
df_gd = df[['Squad', 'GD']].sort_values('GD', ascending=True)

plt.figure(figsize=(8,6))
sns.barplot(
    data=df_gd,
    x='GD',
    y='Squad',
    hue='Squad',
    palette='coolwarm',
    legend=False
)

plt.title("Różnica bramek (GD) – Ekstraklasa 24/25")
plt.xlabel("Różnica bramek")
plt.ylabel("Drużyna")

for i, v in enumerate(df_gd['GD']):
    plt.text(v + (0.5 if v >= 0 else -2.5), i, str(v), va='center', color='black')

plt.axvline(0, color='black', linewidth=1)
plt.tight_layout()
plt.show()
```

```python
# Ranking posiadania piłki
df_poss = df[['Squad', 'Poss']].sort_values('Poss', ascending=True)

plt.figure(figsize=(8,6))
sns.barplot(
    data=df_poss,
    x='Poss',
    y='Squad',
    hue='Squad',
    palette='Purples',
    legend=False
)

plt.title("Posiadanie piłki (%) – Ekstraklasa 24/25")
plt.xlabel("Posiadanie piłki (%)")
plt.ylabel("Drużyna")

for i, v in enumerate(df_poss['Poss']):
    plt.text(v + 0.2, i, f"{v:.1f}", va='center')

plt.tight_layout()
plt.show()
```

```python
# Ranking skuteczności strzałów (gole na strzał)
df_eff = df[['Squad', 'G/Sh']].sort_values('G/Sh', ascending=True)

plt.figure(figsize=(8,6))
sns.barplot(
    data=df_offense,
    x='GF',
    y='Squad',
    hue='Squad', 
    palette='Blues_r',
    legend=False 
)


plt.title("Skuteczność strzałów (Gole na strzał – G/Sh)")
plt.xlabel("Gole na strzał (G/Sh)")
plt.ylabel("Drużyna")

for i, v in enumerate(df_eff['G/Sh']):
    plt.text(v + 0.001, i, f"{v:.3f}", va='center')

plt.tight_layout()
plt.show()
```

```python
# Ranking jakości strzałów celnych (G/SoT)
df_eff_sot = df[['Squad', 'G/SoT']].sort_values('G/SoT', ascending=True)

plt.figure(figsize=(8,6))
sns.barplot(
    data=df_offense,
    x='GF',
    y='Squad',
    hue='Squad',
    palette='Blues_r',
    legend=False
)


plt.title("Skuteczność strzałów celnych (Gole na strzał celny – G/SoT)")
plt.xlabel("Gole na strzał celny (G/SoT)")
plt.ylabel("Drużyna")

for i, v in enumerate(df_eff_sot['G/SoT']):
    plt.text(v + 0.005, i, f"{v:.3f}", va='center')

plt.tight_layout()
plt.show()
```
